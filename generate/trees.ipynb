### 計算數量

```python
with open('tree_data.json', 'r', encoding='utf-8') as f:
	data = json.loads(f.read())
```

```python
data
```

```python
families = {}

for row in data:
	if families.get(row['properties']['tree_type']) == None:
		families[row['properties']['tree_type']] = 1
	else:
		families[row['properties']['tree_type']] += 1

families
```

```python
families_list = [(key, value) for key, value in families.items()]
families_list = sorted(families_list, key=lambda family: family[1])   # sort by count
families_list
```

```python
summary = {
	'data': [
		{
			'name': '河濱生態樹木',
			'data': []
		}
	]
}
for i in range(len(families_list)-1, len(families_list)-1-5, -1):
	summary['data'][0]['data'].append(
		{ 'x': families_list[i][0], 
		  'y': families_list[i][1] } 
	)

summary['data'][0]['data'].append( { 'x': '其他', 'y': 0 } )
for i in range(len(families_list)-1-5, 0, -1):
	summary['data'][0]['data'][-1]['y'] += families_list[i][-1]

summary
```

```python
with open('tree_chart.json', 'w', encoding='utf-8') as f:
	f.write(
		json.dumps(summary, ensure_ascii=False, indent=4)
	)
```
