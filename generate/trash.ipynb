```python
import func
import pandas
import numpy
import json
import requests
```

```python
# trash
trash_r = requests.get('https://data.moenv.gov.tw/api/v2/stat_p_129?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000000&sort=ImportDate%20desc&format=JSON')
trash_json = trash_r.json()

trash_head = dict()
for field in trash_json['fields']:
	trash_head[field['id']] = field['info']['label']

trash_data = pandas.DataFrame(trash_json['records'])
trash_data.rename(columns=trash_head, inplace=True)
trash_data = trash_data[trash_data['統計區'] == '臺北市']
trash_data.insert(1, '年', numpy.nan)
trash_data.insert(2, '月', numpy.nan)
trash_data
```

```python
# recycled
recycled_r = requests.get('https://data.moenv.gov.tw/api/v2/stat_p_132?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000000&sort=ImportDate%20desc&format=JSON')
recycled_json = recycled_r.json()

recycled_head = dict()
for field in recycled_json['fields']:
	recycled_head[field['id']] = field['info']['label']

recycled_data = pandas.DataFrame(recycled_json['records'])
recycled_data.rename(columns=recycled_head, inplace=True)
recycled_data = recycled_data[recycled_data['統計區'] == '臺北市']
recycled_data.insert(1, '年', numpy.nan)
recycled_data.insert(2, '月', numpy.nan)
```

```python
# trash_data
for i in range(len(trash_data)):
	ym = trash_data.iat[i, 0]
	try:
		trash_data.iat[i, 1] = int( ym[:(ym.index('年'))] )
		trash_data.iat[i, 2] = int( ym[(ym.index('年 ')+1):(ym.index('月'))] )
	except ValueError:
		continue
# trash_data[trash_data['月'] > 0]

# recycled_data
for i in range(len(recycled_data)):
	ym = recycled_data.iat[i, 0]
	try:
		recycled_data.iat[i, 1] = int( ym[:(ym.index('年'))] )
		recycled_data.iat[i, 2] = int( ym[(ym.index('年 ')+1):(ym.index('月'))] )
	except ValueError:
		continue
# recycled_data[recycled_data['月'] > 0]
```

```python
trash_result = {
	'data': [
		{
			'name': '廢棄物量', 
			'data': []
		}
	]
}

for month in range(1, 12+1):
	trash_month_data = trash_data[trash_data['月'] == month]
	trash_month_data.describe()
	trash_result['data'][0]['data'].append(
		{
			'y': [trash_month_data.min()['總處理量'], trash_month_data.max()['總處理量']], 
			'x': month
		}
	)

with open('trash_amount_chart.json', 'w', encoding='utf-8') as f:
	trash_output = json.dumps(trash_result, ensure_ascii=False, indent=4)
	f.write(trash_output)
```

```python
recycled_result = {
	'data': [
		{
			'name': '資源回收量', 
			'data': []
		}
	]
}

for month in range(1, 12+1):
	recycled_month_data = recycled_data[recycled_data['月'] == month]
	recycled_month_data.describe()
	recycled_result['data'][0]['data'].append(
		{
			'y': [recycled_month_data.min()['總計'], recycled_month_data.max()['總計']], 
			'x': month
		}
	)

with open('recycled_amount_chart.json', 'w', encoding='utf-8') as f:
	recycled_output = json.dumps(recycled_result, ensure_ascii=False, indent=4)
	f.write(recycled_output)
```
